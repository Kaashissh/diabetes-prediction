```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
print(iris['DESCR'])
```

```python
print(iris['target'])
```

```python
##independent features
X= pd.DataFrame(iris['data'],columns=['sepal length', 'sepal width','petal length','petal width'])
```

```python
##depenedent features
y= iris['target']
```

```python
## train test split 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.2,random_state=10)
```

In decsion tree there is no  need to do standarization and normalization

```python
## apply Decision tree classifier
from sklearn.tree import DecisionTreeClassifier
treeclassifier=DecisionTreeClassifier()
```

```python
treeclassifier.fit(X_train,y_train)
```

```python
## visualize the decision tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
## Here decision tree is splited from top to bottom then there occurs overfiting.
## to reduce overfiting we use post pruning and pre pruning
```

## post pruning 

In post pruning we only take the some part of the decision tree , until its needed. 
for this write max_depth = num in DecisionTreeClassifier.

```python
## apply Decision tree classifier
from sklearn.tree import DecisionTreeClassifier
treeclassifier=DecisionTreeClassifier(max_depth=2)
```

```python
treeclassifier.fit(X_train,y_train)
```

```python
## visualize the decision tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
y_pred=treeclassifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```
